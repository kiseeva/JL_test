## Сводная таблица Комиссия\Рейтинг. Тестовое задание 

Необходимо построить сводную таблицу:
  -  По оси X рейтинг займа
  -  По оси Y комиссия
  -  Сводные данные по сумме займа

```python
import pandas as pd
```

```python
data = pd.read_csv('test.csv')
```

Для составление таблицы нужна предобработка данных.

- В 5 столбцах тип не соответствует данным.
- Есть лишние символы, которые необходимо удалить перед преобразованием типов.

```python
columns = ['Loan issued', 'Comission, %', 'Earned interest',
       'Unpaid,  full amount', 'EL']

for column in columns:
    data[column] = data[column].apply(lambda x: x.translate({ord(i): None for i in ['₽', '%']}))
    data[column] = data[column].apply(lambda x: x.replace(',', '.'))
    data[column] = data[column].apply(lambda x: x.replace(u'\xa0', ''))
    
    data[column] = pd.to_numeric(data[column], errors='coerce')
```

```python
columns = ['Comission, %', 'EL']
data[columns] = data[columns]/ 100
data.head()
```

Составим сводную таблицу и заполним образовавшиеся пропуски.

```python
pivot = data.pivot_table(
    index=['Comission, %'], 
    columns='Рейтинг', 
    values='Loan issued', 
    aggfunc='sum')
```

```python
pivot
```

```python
pivot.fillna(0)
```

В данных по комиссии достаточно большой разброс значений, если данные по каждому значению комиссии не так важны, можно округлить значения до выбранного знака.

```python
data['Comission_round, %'] = round(data['Comission, %'],2)
```

```python
pivot = data.pivot_table(
    index=['Comission_round, %'], 
    columns='Рейтинг', 
    values='Loan issued', 
    aggfunc='sum')
```
